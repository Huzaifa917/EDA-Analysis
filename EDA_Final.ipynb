```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load CSV file
data = pd.read_csv('activity.csv', on_bad_lines='skip')
```

```python
# Access the loaded data
# print the first few rows of the DataFrame
print("First few lines of Data: ")
print(data.head())

# Get the shape of the DataFrame (rows, columns)
print("Shape of Data: ")
print(data.shape)

# Summarize the basic statistics of the data
print("Data Description: ")
print(data.describe())

# Check the data types of each column
print("Data Types: ")
print(data.dtypes)

#Check for missing values
print("Missing values: ")
print(data.isna().head())

#Sum of missing values
print("Missing values sum: ")
print(data.isna().sum())
```

```python
# Now dropping three columns "Sound level, activity and lux since they wont be used in this assignment"
data.drop('lux', axis = 1, inplace = True)
data.drop('soundLevel', axis = 1, inplace = True)
data.drop('activity', axis = 1, inplace = True)
# print the first few rows of the DataFrame
print("First few lines of Data: ")
print(data.head())
```

```python
# There are some missing values as seen by running the function above.
mean_value = data['gZ'].mean()
data['gZ'].fillna(mean_value, inplace=True)
mean_value = data['mX'].mean()
data['mX'].fillna(mean_value, inplace=True)
mean_value = data['mY'].mean()
data['mY'].fillna(mean_value, inplace=True)
mean_value = data['mZ'].mean()
data['mZ'].fillna(mean_value, inplace=True)
```

```python
print(data.describe().T)
```

```python
#Checking for missing values again
#Sum of missing values
print("Missing values sum: ")
print(data.isna().sum())
```

```python
data.fillna(value=0, inplace=True)  # Replace missing values with 0
```

```python
#Checking for missing values again
#Sum of missing values
print("Missing values sum: ")
print(data.isna().sum())
```

```python
correlation_matrix = data.corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='RdBu', center=0, annot_kws={"rotation": 45, "size": 10}, cbar=True, cbar_kws={"orientation": "vertical"})
plt.title('Correlation Matrix')
plt.show()
```

```python
# Calculate the mean of each column
mean_values = data.mean()

# Calculate the median of each column
median_values = data.median()

# Calculate the mode of each column
mode_values = data.mode().iloc[0]

# Calculate the maximum of each column
max_values = data.max()

# Calculate the minimum of each column
min_values = data.min()

# Calculate the standard deviation of each column
std_values = data.std()

# Calculate the variance of each column
var_values = data.var()
```

```python
# Print the calculated statistics for each column
for column in data.columns:
    print(f"Column: {column}")
    print(f"Mean: {mean_values[column]}")
    print(f"Median: {median_values[column]}")
    print(f"Mode: {mode_values[column]}")
    print(f"Maximum: {max_values[column]}")
    print(f"Minimum: {min_values[column]}")
    print(f"Standard Deviation: {std_values[column]}")
    print(f"Variance: {var_values[column]}")
    print()
```

```python
data_without_id = data.drop('_id', axis=1)
# Calculate the mean, median, mode, maximum, minimum, standard deviation, and variance of each column
statistics = pd.DataFrame({
    'Mean': data.mean(),
    'Median': data.median(),
    'Mode': data.mode().iloc[0],
    'Maximum': data.max(),
    'Minimum': data.min(),
    'Standard Deviation': data.std(),
    'Variance': data.var()
})

# Set up the style and context for the plots
sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.0)

# Determine the number of rows and columns for the grid
num_columns = 3
num_rows = (len(data.columns) + num_columns - 1) // num_columns

# Calculate the figure size based on the number of columns
fig_width = num_columns * 6
fig_height = num_rows * 4

# Create the grid of plots
fig, axes = plt.subplots(num_rows, num_columns, figsize=(fig_width, fig_height))

# Iterate over the columns and create individual bar plots in the grid
for i, column in enumerate(data.columns):
    row = i // num_columns
    col = i % num_columns
    ax = axes[row, col]

    sns.barplot(x=statistics.columns, y=statistics.loc[column], palette="viridis", ax=ax)

    # Add labels and title
    ax.set_title(f"Statistics for {column}", fontsize=12)
    ax.set_xlabel("Statistics", fontsize=10)
    ax.set_ylabel("Values", fontsize=10)
    ax.tick_params(axis='x', rotation=45, labelsize=8)
    ax.tick_params(axis='y', labelsize=8)

    # Add data labels to the bars
    for p in ax.patches:
        ax.annotate(f"{p.get_height():.2f}", (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', xytext=(0, 5), textcoords='offset points', fontsize=8)

    # Add black border with 3px margin
    ax.spines['top'].set_visible(True)
    ax.spines['right'].set_visible(True)
    ax.spines['bottom'].set_visible(True)
    ax.spines['left'].set_visible(True)
    ax.spines['top'].set_color('black')
    ax.spines['right'].set_color('black')
    ax.spines['bottom'].set_color('black')
    ax.spines['left'].set_color('black')
    ax.spines['top'].set_linewidth(3)
    ax.spines['right'].set_linewidth(3)
    ax.spines['bottom'].set_linewidth(3)
    ax.spines['left'].set_linewidth(3)
    ax.tick_params(axis='both', which='both', length=0, pad=3)

    # Add padding to the graph
    ax.set_xlim(ax.get_xlim()[0] - 2, ax.get_xlim()[1] + 2)
    ax.set_ylim(ax.get_ylim()[0] - 2, ax.get_ylim()[1] + 2)

# Remove excess subplots if needed
if len(data.columns) < num_rows * num_columns:
    for i in range(len(data.columns), num_rows * num_columns):
        row = i // num_columns
        col = i % num_columns
        fig.delaxes(axes[row, col])

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()

```
